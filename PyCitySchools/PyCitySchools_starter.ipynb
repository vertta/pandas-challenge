# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
student_count = school_data_complete["Student ID"].count()
school_count = len(school_data_complete["school_name"].unique())
```

```python
# Calculate the Totals (Schools and Students)
total_budget = school_data["budget"].sum()
```

```python
# Calculate the Average Scores
average_math_score = school_data_complete["math_score"].mean()
```

```python
average_reading_score = school_data_complete["reading_score"].mean()
```

```python
# Calculate the Percentage Pass Rates
# count the number of student names math score above 70
pass_math_rate=len(school_data_complete.loc[(school_data_complete["math_score"]>= 70)])/student_count * 100
pass_math_rate
```

```python
# count of student names that have reading score above 70
# count of math & read score > 70

pass_reading_rate=len(school_data_complete.loc[(school_data_complete["reading_score"]>= 70)])/student_count * 100 
pass_reading_rate
```

```python
overall_passing_rate = len(school_data_complete.loc[(school_data_complete["math_score"]>= 70)\
                                                     & (school_data_complete["reading_score"]>= 70 )])/student_count * 100  
overall_passing_rate
```

```python
# Minor Data Cleanup
# make a data frame that is a district summary  into one dataFrame
# maybe use mapping to format the total number of students and total budget


# Formatting


# Display the DataFrame
district_summary = pd.DataFrame({
    "Total Schools" : [school_count], # use school count to generate the panda indexes
    "Total Students": student_count,
    "Total Budget": total_budget,
    "Average Math Score":average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math":pass_math_rate,
    "% Passing Reading": pass_reading_rate,
    "% Overall Passing": overall_passing_rate
    
})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.6f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.5f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.6f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.6f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.6f}".format)

district_summary
```

## School Summary

```python
# Determine the School Type
# use set index to take school names and isolate types 
school_types = school_data.set_index(school_data["school_name"])["type"]
school_types.head()


# Calculate the total student count value count of school name column 
counts_per_school= school_data_complete["school_name"].value_counts()
counts_per_school
# Calculate the total school budget and per capita spending, school budget/school counts
budget_per_school = school_data_complete.groupby(["school_name"]).mean()["budget"]
budget_per_school 

capita_per_school = budget_per_school/counts_per_school
capita_per_school

# Calculate the average test scores
average_math_score_per_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_math_score_per_school

average_reading_score_per_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#average_reading_score_per_school 
```

```python
# Get the students who passed math and passed reading by creating separate filtered DataFrames.
school_passing_math = school_data_complete.loc[(school_data_complete["math_score"]>= 70)]
school_passing_math.head()

school_passing_reading = school_data_complete.loc[(school_data_complete["reading_score"]>= 70)]
school_passing_reading.head()
# Get the the students who passed both reading and math in a separate DataFrame.

school_passed_both = school_data_complete.loc[(school_data_complete["math_score"]>= 70)\
                                                     & (school_data_complete["reading_score"]>= 70 )]  
school_passed_both
```

```python
#  Calculate the Percentage Pass Rates
math_per_school_rate = school_passing_math.groupby(["school_name"]).count()["student_name"]/counts_per_school * 100 
math_per_school_rate


reading_per_school_rate = school_passing_reading.groupby(["school_name"]).count()["student_name"]/counts_per_school * 100 
reading_per_school_rate

overall_passing_rate_per_school = school_passed_both.groupby(["school_name"]).count()["student_name"]/counts_per_school * 100 
overall_passing_rate_per_school
```

```python
# Convert to DataFrame
per_school_summary = pd.DataFrame ({
    "School Type ": school_types,
    "Total Students": counts_per_school,
    "Total School Budget": budget_per_school,
    "Per Student Budget": capita_per_school,
    "Average Math Score": average_math_score_per_school,
    "Average Reading Score": average_reading_score_per_school,
    "% Passing Math": math_per_school_rate,
    "% Passing Reading":reading_per_school_rate,
    "% Overall Passing":overall_passing_rate_per_school
})

# Minor data wrangling

#per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:,.6f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:,.6f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:,.6f}".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:,.6f}".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:,.6f}".format)
# Display the DataFrame
per_school_summary
```

## Top Performing Schools (By % Overall Passing)

Math Scores by Grade

```python
# Create data series of scores by grade levels using conditionals


ninethGrader = school_data_complete.loc[(school_data_complete["grade"]=="9th")] #create data series for 9th grade 
tenthGrader = school_data_complete.loc[(school_data_complete["grade"]=="10th")] #create data series for 10th grade 
eleventhGrader = school_data_complete.loc[(school_data_complete["grade"]=="11th")] #create data series for 11th grade 
twelfthGrader = school_data_complete.loc[(school_data_complete["grade"]=="12th")] #create data series for 12th grade 

# Group each by school name
ninethGradeMathScores = ninethGrader.groupby(["school_name"]).mean()["math_score"]
tenthGradeMathScores = tenthGrader.groupby(["school_name"]).mean()["math_score"]
eleventhGradeMathScores = eleventhGrader.groupby(["school_name"]).mean()["math_score"]
twelfthGradeMathScores = twelfthGrader.groupby(["school_name"]).mean()["math_score"]

# Combine series into single DataFrame
scores_by_grade  = pd.DataFrame({
    "9th": ninethGradeMathScores,
    "10th": tenthGradeMathScores,
    "11th": eleventhGradeMathScores,
    "12th": twelfthGradeMathScores   
})

# Minor data wrangling

scores_by_grade["9th"] = scores_by_grade["9th"].map("{:,.6f}".format)
scores_by_grade["10th"] = scores_by_grade["10th"].map("{:,.6f}".format)
scores_by_grade["11th"] = scores_by_grade["11th"].map("{:,.6f}".format)
scores_by_grade["12th"] = scores_by_grade["12th"].map("{:,.6f}".format)

# remove the index name 
scores_by_grade.index.name = None

# Display the DataFrame
scores_by_grade
```
